```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://mcdindia.com/#menu"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
print(response.text[:1000])
```

```python
# Headers to mimic a browser request
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

# Send an HTTP GET request
response = requests.get(url, headers=headers)

soup = BeautifulSoup(response.text, 'html.parser')
response
```

```python
print (soup.prettify()) 
# print the HTML content of the page 
```

```python
i = soup.find_all('div', class_ = "slide-heading sub-heading")
```

```python
name = []
for element in i:
    name.append(element.text.strip())
name
```

```python
ss = soup.select('p.slide-serving-value')
ss_text = [s.text.strip() for s in ss]
ss_text
```

```python
alg = soup.find_all('p', class_="slide-allergen-value")
len_alg = len(alg)
len_alg
```

```python
for element in alg:
    print(element)
```

```python
Calories= []

nute = soup.find_all('div', class_ = "nutrition-value value-1 para-text")
for nute in nute:
    Calories.append(nute.text.strip())


Calories
```

```python
Calories = [item.replace('Energy\n', '') for item in Calories]
Calories_in_Kcal = [item.replace('kCal', '') for item in Calories]
Calories_in_Kcal = [item.replace('\n', '|') for item in Calories_in_Kcal]
len(Calories_in_Kcal)
```

```python
results_dict = {}
for i in range(1,11):
    auto = soup.find_all('div', class_ = f"nutrition-value value-{i} para-text")
    result_list = [item.get_text() for item in auto]
    
    # Append the list to the dictionary with a unique key
    results_dict[f"value_{i}"] = result_list

# Print the results
for key, result_list in results_dict.items():
    print(f"{key}: {result_list}")
    
value_2_results = results_dict.get("value_2", [])
value_3_results = results_dict.get("value_3", [])
value_4_results = results_dict.get("value_4", [])
value_5_results = results_dict.get("value_5", [])
value_6_results = results_dict.get("value_6", [])
value_7_results = results_dict.get("value_7", [])
value_8_results = results_dict.get("value_8", [])
value_9_results = results_dict.get("value_9", [])
value_10_results = results_dict.get("value_10", [])
```
